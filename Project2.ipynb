```python
import csv
import matplotlib.pyplot as plt
from scipy import stats

with open('NY_2006.tab') as file6:
    file6_reader = csv.reader(file6, delimiter='\t')
    list6 = list(file6_reader)
    
with open('NY_2008.tab') as file8:
    file8_reader = csv.reader(file8, delimiter='\t')
    list8 = list(file8_reader)
    
with open('NY_2010.tab') as file10:
    file10_reader = csv.reader(file10, delimiter='\t')
    list10 = list(file10_reader)
    
#average temperatures on election day
t6 = 44
t8 = 59
t10 = 49
t = [t6,t8,t10]
```

```python
#2006
votes6 =[[0 for x in range(int(list6[-1][3]))] for x in range(4)]

for i in range(1,len(votes6[0])+1):
    votes6[0][i-1] = i

    
for j in range(int(list6[-1][3])):
    for i in range(1,len(list6)):
        if int(list6[i][3]) == votes6[0][j-1]:
            votes6[1][j-1] += float(list6[i][53])
            votes6[2][j-1] += float(list6[i][54])
            votes6[3][j-1] += float(list6[i][55])
 


#2008
votes8 =[[0 for x in range(int(list8[-1][3]))] for x in range(4)]


for i in range(1,len(votes8[0])+1):
    votes8[0][i-1] = i

    
for j in range(int(list8[-1][3])):
    for i in range(1,len(list8)):
        if int(list8[i][3]) == votes8[0][j-1] and list8[i][49] != '':
            votes8[1][j-1] += float(list8[i][49])
            votes8[2][j-1] += float(list8[i][50])
            votes8[3][j-1] += float(list8[i][51])

            
#2010
votes10 =[[0 for x in range(int(list10[-1][3]))] for x in range(4)]

for i in range(1,len(votes10[0])+1):
    votes10[0][i-1] = i

for j in range(int(list10[-1][3])):
    for i in range(1,len(list10)):
        if int(list10[i][3]) == votes10[0][j-1]:
            votes10[1][j-1] += float(list10[i][53])
            votes10[2][j-1] += float(list10[i][54])
            votes10[3][j-1] += float(list10[i][55])
```

```python
total6 = [0 for x in range(3)]
for i in range(3):
    total6[i] = sum(votes6[i+1])
    
    
total8 = [0 for x in range(3)]
for i in range(3):
    total8[i] = sum(votes8[i+1])
    
    
    
total10 = [0 for x in range(3)]
for i in range(3):
    total10[i] = sum(votes10[i+1])

    
p6 = [0 for x in range(2)]
p6[0] = total6[1]/total6[0]
p6[1] = total6[2]/total6[0]

p8 = [0 for x in range(2)]
p8[0] = total8[1]/total8[0]
p8[1] = total8[2]/total8[0]

p10 = [0 for x in range(2)]
p10[0] = total10[1]/total10[0]
p10[1] = total10[2]/total10[0]
```

```python
pd = [p6[0],p8[0],p10[0]]
slope, intercept, r_value, p_value, std_err = stats.linregress(t,pd)
plot2 = [t[0]*slope+intercept,t[1]*slope+intercept,t[2]*slope+intercept]

pr = [p6[1],p8[1],p10[1]]
slope, intercept, r_value, p_value, std_err = stats.linregress(t,pr)
plot3 = [t[0]*slope+intercept,t[1]*slope+intercept,t[2]*slope+intercept]

dp = [p6[0]-p6[1],p8[0]-p8[1],p10[0]-p10[1]]
slope, intercept, r_value, p_value, std_err = stats.linregress(t,dp)
plot1 = [t[0]*slope+intercept,t[1]*slope+intercept,t[2]*slope+intercept]
t = [t6,t8,t10]
plt.plot(t,pd,'bo')
plt.plot(t,pr,'ro')
plt.plot(t,dp,'go')
plt.plot(t,plot1,'g')
plt.plot(t,plot2,'b')
plt.plot(t,plot3,'r')
plt.show()
print('The blue line represents votes for Democrats, the red Republican, and the green is the difference (in percentage).')
print('The slope of the green line is',slope*100,'%.')
```
